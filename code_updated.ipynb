```python
# Importing numpy library
import numpy as np

# Replacing non-standard data with NaN
data = data.replace( "?" , np.NaN)

# Checking if the values are replaced correctly
data.tail(20)
```

```python
# Replacing NaN values in numeric variables with mean values for that variable
data.fillna(data.mean(), inplace=True)

# Let us check if the numeric variables have any missing values
data.isna().sum()
```

```python
#the character variables still have NaN values
data.tail(20)
```

```python
# Creating a loop to run on character variables to replace NaN values with most frequent values in that variable
for i in data:
    # if the variable is object type
    if data[i].dtypes == 'object':
        # replace with the most frequent value
        data = data.fillna(data[i].value_counts().index[0])

# Let us check if any variables have any missing values
data.isna().sum()
```

```python
#Converting non-numeric data into numeric data using LabelEncoder

from sklearn import preprocessing

encode = preprocessing.LabelEncoder()

#extracting data types for each column and performing numeric transformation
for i in data:
    if data[i].dtypes =='object':
        data[i]=encode.fit_transform(data[i])
```

```python
#lets check if transformation has worked correctly
data.tail(20)
```

```python
#creating correlation matrix for our data to check correlation of variables and possibility of multicollinearity
corr = data.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#We can observe feature 4,3 and 9,10 are highly correlated, this might lead to multicollinearity
#we can also observe that feature 0,6,11 have very less correlation with variable 15 which is our dependent variable

# So we decide to drop features 0,4,6,9,11 and convert the DataFrame to a NumPy array

data = data.drop([0,4,6,9,11],axis=1)

corr = data.corr()

import matplotlib.pyplot as plt
import seaborn as sns

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

```

```python
#Now we will scale the data using MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

data = data.values

# We will split independent variables(x) and dependent variable(y)
x = data[:,0:9] 
y = data[:,10]

#Scaling the data
scale = MinMaxScaler(feature_range=(0,1))
new_x = scale.fit_transform(data)
```

```python
#let check how our data looks like now
print(data)
```

```python
# Now we will split the data into train(%) and test(%) set to build a model

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.40)
```

```python
#Now we will build our first model using KNN

from sklearn.neighbors import KNeighborsClassifier  

kmodel = KNeighborsClassifier(n_neighbors=5)  
kmodel.fit(x_train, y_train) 
```

```python
# To check our model performance, we will check the confusion matrix ie, True +ve,-ve and False +ve,-ve

from sklearn.metrics import confusion_matrix

predicted = kmodel.predict(x_test)
modelscore = kmodel.score(x_test,y_test)

print("Accuracy: ", modelscore)

print(confusion_matrix(y_test,predicted))
```

```python
#Now we will build our second model using Logistic Regression

from sklearn.linear_model import LogisticRegression

# Initialize a Logistic Regression classifier and fit model on train set

lmodel = LogisticRegression()

lmodel.fit(x_train,y_train)
```

```python
# Now again we predict the values of dependent variable using new model and check model accuracy
predicted = lmodel.predict(x_test)

print("Accuracy: ", lmodel.score(x_test,y_test))

print(confusion_matrix(y_test,predicted))
```
